# Data Analyzing And Visualization 

# Analyze The Coloumn 'churn'

```python
data.head()
```

```python
data['churn'].unique()
```

```python
A = data['churn'].value_counts()
A
```

# Donout Chart 

```python
plt.pie(A,labels=("Not Churn","Churn"),startangle= 50,shadow= True , explode=(0,0.15),autopct="%1.2f%%",pctdistance=(0.75))
Circle = plt.Circle((0,0),0.5,color='white')
plt.gcf
plt.gca().add_artist(Circle)
plt.title('Churned Vs Non Churned',weight = 'bold')
```

# Bar Plot 

```python
sns.barplot(A,palette="spring",)
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.set(style='whitegrid')
sns.countplot(data, x="state",palette='spring',legend=True,hue=data.churn)
plt.xticks(rotation = 90)
```

```python
data.head()
```

```python
gb =data.groupby("state").agg({"total night calls":"mean","total night minutes":"mean"})
gb
```

```python
plt.figure(figsize=(12,6))
plt.xticks(rotation=90)
sns.scatterplot(gb,palette="spring")
plt.plot(gb)
plt.title("Total Night Calls And Minutes By State", weight= 'bold')
```
